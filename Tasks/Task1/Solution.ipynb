## 4. There are some empty cells (called missing data) in the dataset. Fill them with mean of the corresponding column.

let's see how many NaN values exists in our DataFrame:

```python
df.isnull().sum()
```

As you can see, BMI column has some missing values. Let's see which feature correlates more with the BMI feature :

```python
df.corr()['BMI'].drop('BMI').sort_values(ascending = False)
```

It looks like SkinThickness correlates more this BMI. We will group the dataframe by sickthickness and calculate the mean value for BMI per SkinThickness:

```python
BMI_avg_per_SkinThickness = df.groupby('SkinThickness').mean()['BMI']
BMI_avg_per_SkinThickness
```

```python
def fill_BMI(skinthickness,bmi) :
    if np.isnan(bmi) :
        return BMI_avg_per_SkinThickness[int(skinthickness)]
    else :
        return bmi
```

```python
df['BMI'] = df.apply(lambda row : fill_BMI(row['SkinThickness'],row['BMI']),axis = 1)
```

```python
df.isnull().sum()
```
